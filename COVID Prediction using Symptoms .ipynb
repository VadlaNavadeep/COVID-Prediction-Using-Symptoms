```python
# Step 1: Import required libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Step 2: Prepare a simple dataset
# 'Yes' and 'No' will be mapped to 1 and 0 for model training
data = {
    'Fever':     ['Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No'],
    'Cough':     ['Yes', 'No',  'Yes','Yes', 'No', 'Yes', 'Yes', 'No'],
    'Headache':  ['Yes', 'No',  'No', 'Yes', 'No', 'No',  'Yes','No'],
    'COVID':     ['Yes', 'No',  'No', 'Yes', 'No', 'Yes', 'Yes','No']
}

df = pd.DataFrame(data)

# Step 3: Convert Yes/No to 1/0
df.replace({'Yes': 1, 'No': 0}, inplace=True)

# Step 4: Split into features (X) and target (y)
X = df[['Fever', 'Cough', 'Headache']]
y = df['COVID']

# Step 5: Train the Decision Tree model
model = DecisionTreeClassifier()
model.fit(X, y)

# Step 6: Visualize the decision tree
plt.figure(figsize=(10,6))
tree.plot_tree(model, feature_names=['Fever', 'Cough', 'Headache'], 
               class_names=['No', 'Yes'], filled=True)
plt.title("Decision Tree - COVID Prediction")
plt.show()

# Step 7: Predict for a new person
# Example: Fever = Yes, Cough = No, Headache = Yes  [1, 0, 1]
sample = [[1, 0, 1]]
prediction = model.predict(sample)
print("COVID Prediction (1 = Yes, 0 = No):", prediction[0])

```
